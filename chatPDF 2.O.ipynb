```python
import os 

from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from sentence_transformers import SentenceTransformer
```

```python
#first step is to load a document
loader = PyPDFLoader(r"C:\Users\MOIZ\Desktop\AI_in_2026.pdf")
documents=loader.load()
```

```python
len(documents)
```

```python
#second step to make chunks
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
texts = text_splitter.split_documents(documents)
```

```python
len(texts)
```

```python
texts = [doc.page_content for doc in documents]

```

```python
print(texts)
```

```python
#create embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
vector=embeddings.embed_documents(texts)
```

```python
print(len(vector))
```

```python
#create vector store
db=Chroma.from_documents(documents,embeddings)
```

```python
retriever = db.as_retriever()
```

```python
# expose this index in a retriever interface
retriever = db.as_retriever(
    search_type="similarity", search_kwargs={"k": 3}
)
```

```python
import re
import panel as pn
from langchain_core.prompts import ChatPromptTemplate

def md_to_html(text):
    """Convert basic markdown to HTML."""
    text = re.sub(r'\*\*(.+?)\*\*', r'<strong>\1</strong>', text)  # **bold**
    text = re.sub(r'\*(.+?)\*',     r'<em>\1</em>',           text)  # *italic*
    text = re.sub(r'`(.+?)`',       r'<code>\1</code>',        text)  # `code`
    text = re.sub(r'\n- ',          r'<br>• ',                 text)  # bullet lists
    text = text.replace('\n', '<br>')
    return text

pn.extension(sizing_mode="stretch_width")

# ── Widgets ───────────────────────────────────────────────────────────────────
file_input = pn.widgets.FileInput(accept=".pdf", name="Upload PDF")

process_button = pn.widgets.Button(
    name="⚙ Process PDF", button_type="success", width=160, height=40
)

status = pn.pane.Alert(
    "Upload a PDF and click **Process PDF** to get started.",
    alert_type="info", sizing_mode="stretch_width",
)

question_input = pn.widgets.TextInput(
    placeholder="Type your question here…", sizing_mode="stretch_width"
)

ask_button = pn.widgets.Button(
    name="Ask ➤", button_type="primary", width=100, height=40
)

chat_area = pn.pane.HTML(
    "", sizing_mode="stretch_width",
    styles={
        "background": "#f8f9fa", "border": "1px solid #dee2e6",
        "border-radius": "8px", "padding": "16px",
        "min-height": "300px", "font-size": "14px",
    },
)

# ── State ─────────────────────────────────────────────────────────────────────
_db = None
_retriever = None
_chat_history = []

# Reuse the prompt template from your notebook
_prompt_template = ChatPromptTemplate.from_template(
    "Answer the question based only on the context.\n\n{context}\n\nQuestion: {question}"
)

# ── Callbacks ─────────────────────────────────────────────────────────────────
def process_pdf(event):
    global _db, _retriever

    if not file_input.value:
        status.object = "⚠ Please select a PDF first."
        status.alert_type = "warning"
        return

    process_button.disabled = True
    status.object = "⏳ Processing PDF…"
    status.alert_type = "info"

    try:
        pdf_path = "uploaded.pdf"
        with open(pdf_path, "wb") as f:
            f.write(file_input.value)

        from langchain_community.document_loaders import PyPDFLoader
        from langchain_text_splitters import RecursiveCharacterTextSplitter

        loader = PyPDFLoader(pdf_path)
        documents = loader.load()

        splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
        split_docs = splitter.split_documents(documents)

        _db = Chroma.from_documents(split_docs, embeddings)
        _retriever = _db.as_retriever(search_type="similarity", search_kwargs={"k": 3})

        status.object = "🎉 Processing complete! Your PDF is ready — go ahead and ask questions below."
        status.alert_type = "success"
        # Flash the chat card border to draw attention
        chat_card.styles = {
            "background": "#fff", "border": "2px solid #16a34a",
            "border-radius": "10px", "padding": "20px",
        }

    except Exception as e:
        status.object = f"❌ Error: {e}"
        status.alert_type = "danger"
    finally:
        process_button.disabled = False


def ask_question(event):
    global _retriever, _chat_history

    if _retriever is None:
        status.object = "⚠ Please upload and process a PDF first."
        status.alert_type = "warning"
        return

    question = question_input.value.strip()
    if not question:
        return

    ask_button.disabled = True
    question_input.disabled = True

    try:
        docs = _retriever.invoke(question)
        context = "\n\n".join(doc.page_content for doc in docs)

        # Use the same prompt template as your notebook
        formatted_prompt = _prompt_template.format(context=context, question=question)
        response = chat_model.invoke(formatted_prompt)  # uses chat_model from your notebook
        answer = response.content if hasattr(response, "content") else str(response)

        answer = md_to_html(answer.strip())

        _chat_history.append(
            f'<div style="background:#dbeafe;border-radius:6px;padding:8px 12px;margin-bottom:6px">'
            f'<strong>🧑 You:</strong> {question}</div>'
        )
        _chat_history.append(
            f'<div style="background:#f0fdf4;border-radius:6px;padding:8px 12px;margin-bottom:6px">'
            f'<strong>🤖 AI:</strong><br>{answer.replace(chr(10), "<br>")}</div>'
        )
        chat_area.object = "<br>".join(_chat_history)
        question_input.value = ""

    except Exception as e:
        status.object = f"❌ Error: {e}"
        status.alert_type = "danger"
    finally:
        ask_button.disabled = False
        question_input.disabled = False


process_button.on_click(process_pdf)
ask_button.on_click(ask_question)

# ── Layout ────────────────────────────────────────────────────────────────────
upload_card = pn.Column(
    pn.pane.Markdown("### 📄 Upload Document", margin=(0, 0, 4, 0)),
    file_input,
    pn.Row(process_button, align="end"),
    status,
    styles={"background": "#fff", "border": "1px solid #dee2e6",
            "border-radius": "10px", "padding": "20px"},
    sizing_mode="stretch_width",
)

chat_card = pn.Column(  # NOTE: referenced in process_pdf callback above
    pn.pane.Markdown("### 💬 Chat", margin=(0, 0, 4, 0)),
    chat_area,
    pn.Row(question_input, ask_button, sizing_mode="stretch_width"),
    styles={"background": "#fff", "border": "1px solid #dee2e6",
            "border-radius": "10px", "padding": "20px"},
    sizing_mode="stretch_width",
)

app = pn.Column(
    pn.pane.Markdown(
        "# 🔍 PDF RAG Chatbot",
        styles={"color": "#1e3a5f", "border-bottom": "2px solid #1e3a5f", "padding-bottom": "8px"},
    ),
    upload_card,
    chat_card,
    max_width=860,
    margin=20,
    styles={"margin-left": "auto", "margin-right": "auto"},
)

app
```
